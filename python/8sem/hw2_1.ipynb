```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline

a = 1.
Nknots = 101
N = 100
V = 100

greed = np.empty((Nknots,Nknots), [('x',np.float64), ('y',np.float64)])
greed['x'], greed['y'] = np.meshgrid(np.linspace(0, a, Nknots), np.linspace(0, a, Nknots))

u = np.zeros(shape = (Nknots, Nknots))
```

```python
for k in range(N):
    u += 4*V/np.pi*np.sinh(np.pi*(a - greed['x'])/a*(2*k + 1)) * np.sin(np.pi*(greed['y'])/a*(2*k + 1)) \
                                                      /( (2*k+1) * np.sinh(np.pi*(2*k + 1)) )
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(greed['x'], greed['y'], u, rstride=5, cstride=5)
```
